```python
import seaborn as sns, pandas as pd
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import ListedColormap
from sklearn.preprocessing import MinMaxScaler
from scipy import stats
import numpy as np
```

## Iteração 500

* BETA  = 10
* GAMMA = 10

```python
dataframe = pd.read_csv('result_5_10_10.csv', sep=',')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Suponha que já tem um dataframe com as colunas corretas
filenames   = dataframe['Filename']
oz_sizes    = dataframe['cp_oz']
model_sizes = dataframe['cp_model']
cmp_sizes   = dataframe['cmp_cp_model']

x = np.arange(len(filenames))  # índice para cada barra
width = 0.25  # espaço de cada barra

fig, ax = plt.subplots(figsize=(12, 6))

# Três grupos de barras, cada um com deslocamento diferente
bars1 = ax.bar(x - width, model_sizes, width, label='Model compilation time', color='royalblue')
bars2 = ax.bar(x, oz_sizes, width, label='-Oz compilation time', color='orange')
bars3 = ax.bar(x + width, cmp_sizes, width, label='Cmp compilation time', color='purple')

# Personalizações
ax.set_xlabel('Benchmark name')
ax.set_ylabel('Compilation time (ms)')
ax.set_title('Comparison of Compilation Times (alpha=5, beta=10, gamma=10)')
ax.set_xticks(x)
ax.set_xticklabels(filenames, rotation=45, ha='right')
ax.legend()

plt.tight_layout()
plt.show()
```
